```python
from sudoku import solver
from sudoku.sudoku import Sudoku
from sudoku.renderer import TextRenderer
```

```python
r = TextRenderer()

with open('p096_sudoku.txt') as infile:
    puzzles = []
    for p in infile.read().split('Grid'):
        text_form = p.strip()[3:].replace('\n', '|').replace('0', '.')
        if not text_form: continue
        print(text_form)
        puzzles.append(Sudoku.from_text(text_form))
```

```python
print(r.render(puzzles[0]))
```

```python
s = solver.Solver(puzzles[0], verbose=False)
s.solve()
print(r.render(s.sudoku))
```

```python
total = 0
for i, p in enumerate(puzzles):
    s = solver.Solver(p, verbose=False)
    assert s.solve(), f'puzzle {i} could not be solved'
    total += int(''.join(s.sudoku.at(cell) for cell in ('A1', 'A2', 'A3')))
```
